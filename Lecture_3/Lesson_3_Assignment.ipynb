# Graph
Graph: A graph is a set of vertices (or nodes) and a set of edges. Each edge connects a pair of vertices. In the example provided, we have a graph with vertices ‘X’, ‘Y’, ‘Z’, and ‘F’, and edges connecting ‘X’ and ‘Z’, ‘Y’ and ‘X’, and ‘Z’ and ‘Y’.

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.Graph()
G.add_node('F')
G.add_edges_from([('X','Z'),('Y','X'),('Z','Y')])

pos = nx.spring_layout(G)

nx.draw_networkx_nodes(G,pos, node_size=500)
nx.draw_networkx_edges(G,pos)
nx.draw_networkx_labels(G,pos)
plt.show()

```

# Directed Graphs

Directed Graphs: A directed graph, or digraph, is a graph where each edge has a direction. In the example, the edges point from ‘X’ to ‘Z’, from ‘Y’ to ‘X’, and from ‘Z’ to ‘Y’.

```python
G = nx.DiGraph()
G.add_edges_from([('X','Z'),('Y','X'),('Z','Y')])

pos = nx.circular_layout(G)

nx.draw_networkx_nodes(G,pos, node_size=500)
nx.draw_networkx_edges(G,pos, edgelist=G.edges(), edge_color='black')
nx.draw_networkx_labels(G,pos)
plt.show()
```

# Simple Graphs

Simple Graphs: A simple graph is a graph with no loops (edges connected at both ends to the same vertex) and no more than one edge between any two different vertices. The example shows a simple graph with an additional vertex ‘W’ connected to ‘X’.

```python
G = nx.Graph()
G.add_edges_from([('X','Z'),('Y','X'),('Z','Y'), ('X', 'W')])

pos = nx.shell_layout(G)

nx.draw_networkx_nodes(G,pos, node_size=500)
nx.draw_networkx_edges(G,pos, edgelist=G.edges(), edge_color='black')
nx.draw_networkx_labels(G,pos)
plt.show()
```

# Complete Graphs

Complete Graphs: A complete graph is a simple graph in which every pair of distinct vertices is connected by a unique edge. In the example, we have a complete graph with vertices ‘X’, ‘Y’, ‘Z’, ‘P’, ‘Q’, ‘R’, ‘S’, and ‘T’.

```python
G = nx.Graph()
G.add_edges_from([('X','Y'),('X','Z'),('Y','Z'),('P','Q'), ('R', 'S')])  
G.add_node ('T')

pos = nx.kamada_kawai_layout(G)

nx.draw_networkx_nodes(G, pos, node_size=500)
nx.draw_networkx_edges(G, pos, edgelist=G.edges(), edge_color='black')
nx.draw_networkx_labels(G, pos)
plt.show()
```

# Bipartite Graphs

Bipartite Graphs: A bipartite graph is a graph whose vertices can be divided into two disjoint sets such that every edge connects a vertex in one set to a vertex in the other set. The example shows a bipartite graph with vertices {1,2,3} in one set and {4,5,6} in the other.

```python
G = nx.Graph()
G.add_nodes_from([1,2,3], bipartite=0) 
G.add_nodes_from([4,5,6], bipartite=1)
G.add_edges_from([(1,4), (1,5), (1,6)
                ,(2,4), (2,5), (2,6), (3,4)])

top = nx.bipartite.sets(G)[0]
pos = nx.bipartite_layout(G, top)
nx.draw(G, pos=pos, with_labels=True, node_color=['pink','magenta','red','deepskyblue','cyan', 'purple'])
plt.show()
```

# Subgraphs

Subgraphs: A subgraph is a portion of a graph, which may include some vertices and edges of the graph. In the example, graph ‘H’ is a subgraph of graph ‘G’.

```python
G = nx.Graph()
H = nx.Graph()
G.add_edges_from([('X','Y'),('X','Z'),('Z','P'),('Y','P'),('X','P'),('Y','Z'), ('W', 'X')])
H.add_edges_from([('X','Y'),('Y','P'),('P','X'), ('W', 'X')])

pos = {'X': (0, 0), 'Y': (-1, 1), 'P': (0, 2), 'Z': (1, 1), 'W': (2, 2)}

plt.figure(figsize=(10, 5))  

plt.subplot(1, 2, 1)  
nx.draw_networkx_nodes(G, pos, node_size=500)
nx.draw_networkx_edges(G, pos, edgelist=G.edges(), edge_color='black')
nx.draw_networkx_labels(G, pos)
plt.title('Graph G')
plt.axis('off')  

plt.subplot(1, 2, 2)  
nx.draw_networkx_nodes(H, pos, node_size=500)
nx.draw_networkx_edges(H, pos, edgelist=H.edges(), edge_color='black')
nx.draw_networkx_labels(H, pos)
plt.title('Graph H')
plt.axis('off')  

plt.tight_layout()  
plt.show()
```

# Degree of Vertex

Degree of Vertex: The degree of a vertex in a graph is the number of edges incident to it.

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.Graph()
G.add_edges_from([('Y','P'),('Y','X'),('P','P'), ('W', 'Y')])  

pos = nx.planar_layout(G)

nx.draw_networkx_nodes(G, pos, node_size=500)
nx.draw_networkx_edges(G, pos, edgelist=G.edges(), edge_color='black')
nx.draw_networkx_labels(G, pos)
plt.show()

# Print the degree of each vertex
for node in G.nodes():
    print(f"The degree of vertex {node} is {G.degree(node)}")
```

# Handshake Theorem

The Handshaking Theorem is a fundamental concept in graph theory. It states that the sum of the degrees of all vertices in a graph is equal to twice the number of edges. This is because each edge contributes exactly two to the total degree count - one for each endpoint.

Mathematically, it can be represented as:

$$\sum_{v \in V} deg(v) = 2|E|$$

where:
- `V` is the set of vertices
- `deg(v)` is the degree of vertex `v`
- `E` is the set of edges

## Example

Consider a simple undirected graph with 5 vertices and 4 edges:


The degrees of the vertices are as follows:
- Vertex 1: degree 2
- Vertex 2: degree 3
- Vertex 3: degree 3
- Vertex 4: degree 2
- Vertex 5: degree 1

Adding up the degrees gives us a total of 11. There are 4 edges in the graph, and 2*4 = 8. So, in this case, the Handshaking Theorem does not hold true.

However, if we consider each edge as two separate half-edges (one at each vertex it connects), we can see that the total number of half-edges is indeed 11, which matches the total degree count. This is the essence of the Handshaking Theorem.

# Travelling in a Graph

1. **Walk**: A walk in a graph is a sequence of vertices and edges, where each edge's endpoints are the preceding and following vertices in the sequence. For example, in a graph with vertices A, B, C, and D, a walk could be A-B-C-D-B. This walk includes the vertices A, B, C, D and the edges AB, BC, CD, and DB.

    $$W = A - B - C - D - B$$

2. **Closed Walk**: A closed walk is a walk that starts and ends at the same vertex. For example, in a graph with vertices A, B, C, and D, a closed walk could be A-B-C-A. This walk starts and ends at vertex A.

    $$CW = A - B - C - A$$

3. **Trail**: A trail is a walk in which no edges are repeated. For example, in a graph with vertices A, B, C, and D, a trail could be A-B-C-D. This trail includes the vertices A, B, C, D and the edges AB, BC, and CD, none of which are repeated.

    $$T = A - B - C - D$$

4. **Path**: A path is a walk in which no vertices are repeated. For example, in a graph with vertices A, B, C, and D, a path could be A-B-C-D. This path includes the vertices A, B, C, D, none of which are repeated.

    $$P = A - B - C - D$$

5. **Circuit**: A circuit is a closed walk that includes at least one edge and does not repeat any edges. For example, in a graph with vertices A, B, C, and D, a circuit could be A-B-C-D-A. This circuit starts and ends at vertex A and includes the edges AB, BC, CD, and DA, none of which are repeated.

    $$C = A - B - C - D - A$$

6. **Simple Circuit**: A simple circuit is a circuit that does not pass through the same vertex more than once (except for the starting and ending vertex, which are the same). For example, in a graph with vertices A, B, C, D, and E, a simple circuit could be A-B-C-D-E-A. This simple circuit starts and ends at vertex A and includes the edges AB, BC, CD, DE, and EA, none of which are repeated.

    $$SC = A - B - C - D - E - A$$

# Euler's Circuit

Euler’s Circuit: An Euler’s circuit in a graph is a route that passes through every edge exactly once and returns to the starting vertex. The example shows a graph where such a circuit is possible.

```python
G = nx.Graph()
for node in ['X', 'Y', 'Z', 'P', 'W']:
    G.add_node(node)

edges = [('X', 'Y'), ('Y', 'Z'), ('Z', 'P'), ('P', 'X'), ('W', 'X')]

for edge in edges:
    G.add_edge(*edge)

nx.draw(G,with_labels=True, node_color='lightblue', edge_color='gray')
plt.show()
```

# Hamiltonian Circuit

Hamiltonian Circuits: A Hamiltonian circuit in a graph is a route that visits every vertex exactly once and returns to the starting vertex. The example shows a graph where such a circuit is possible.

```python
G=nx.Graph()

G.add_nodes_from("ABCDE")
G.add_edges_from(["AC","AD","AB","CD","CE","CB","DE", "EA"])


pos={
    "A":(0,2),
    "B":(2,2),
    "C":(1,1),
    "D":(0,0),
    "E":(2,0)
}

nx.draw(G,pos=pos,with_labels=True,node_color="lightblue",node_size=2000,font_size=20,width=1)
plt.margins(0.2)
plt.show()
```
