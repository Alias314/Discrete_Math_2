**Question 1:**

The expected value of a random variable is the sum of all possible values it can take, each multiplied by the probability of that outcome. For a multinomial distribution, the expected value of $Y_i$ (the number of trials that result in outcome $i$) is given by $E[Y_i] = n \cdot p_i$.

The expected value $E[Y_i]$ is defined as:

\begin{equation}
E[Y_i] = \sum_{y=0}^{n} y \cdot P(Y_i = y)
\end{equation}

For a multinomial experiment, $P(Y_i = y)$ is given by the multinomial probability formula. However, because $Y_i$ can only take on the values 0 through $n$, and the probability of any other outcome is 0, we can simplify this to:

\begin{equation}
E[Y_i] = \sum_{y=0}^{n} y \cdot \binom{n}{y} p_i^y (1-p_i)^{n-y}
\end{equation}

This is the formula for the expected value of a binomial distribution, which simplifies to $n \cdot p_i$. Therefore, we have:

\begin{equation}
E[Y_i] = n \cdot p_i
\end{equation}

**Question 2:**

Given the proportions for each age group, we can calculate the probability of getting a certain number of people in each age group using the multinomial distribution formula. The expected value for obtaining a person in the age group 65 and above can be calculated using the formula for the expected value of a multinomial distribution.

The probability $P$ of getting 100 people between 18 and 24, 200 people between 25 and 34, and 200 people between 45 and 64 is given by:

\begin{equation}
P = \frac{500!}{100!200!200!} \cdot (0.18)^{100} \cdot (0.23)^{200} \cdot (0.27)^{200}
\end{equation}

The expected value $E$ for obtaining a person in the age group 65 and above is given by:

\begin{equation}
E = 500 \cdot 0.16 = 80
\end{equation}

Therefore, the expected value for to obtain a person in the age group 65 and above is 80

**Question 1: Derive the probability distribution of the experiment**

Given an experiment with $n$ trials, each trial can result in either success or failure.

The binomial distribution gives us the probability of observing a specific number of successes in a fixed number of trials. In this case, we want to find the probability of observing $k$ successes.

The formula for the binomial distribution is:

$$ P(Y = k) = \binom{n}{k} p^k (1 - p)^{n - k} $$

Here, $n$ represents the total number of trials, $p$ is the probability of success in a single trial, and $k$ is the number of successes we are interested in.

This formula gives us the probability distribution for the experiment.

**Question 2: Show that the expectation of this probability distribution is $E[y] = \Sigma y \cdot P(y) = np$**

The expected value $E[Y]$ of a binomial distribution is given by $np$, where $n$ is the number of trials and $p$ is the probability of success on a single trial. This can be derived from the definition of expected value:

$$ E[Y] = \Sigma y \cdot P(Y = y) $$

Substituting the PMF of the binomial distribution:

$$ E[Y] = \Sigma y \cdot \binom{n}{y} p^y (1-p)^{n-y} $$

After simplifying, we get:

$$ E[Y] = np $$

**Question 3: What is the probability that at least nine of ten receiving the medication will recover?**

Here, $n = 10$, $p = 0.30$ (probability of recovery), and we want to find $P(Y \geq 9)$.

We can use the PMF of the binomial distribution to calculate this:

$$ P(Y \geq 9) = P(Y = 9) + P(Y = 10) = \binom{10}{9} (0.30)^9 (0.70) + \binom{10}{10} (0.30)^{10} $$

Calculating the above expression:

$$ P(Y \geq 9) = 10 \times (0.30)^9 \times (0.70) + 1 \times (0.30)^{10} $$

$$ P(Y \geq 9) = 0.001590386 + 0.0000059049 $$

$$ P(Y \geq 9) = 0.001596291 $$

So, the probability that at least nine of ten receiving the medication will recover is approximately 0.0016 or 0.16%.

**Question 4: Find the probability of observing at least one defective**

Here, $n = 5$, $p = 0.05$ (probability of a fuse being defective), and we want to find $P(Y \geq 1)$.

It's easier to find the probability of no defective fuses and subtract it from 1:

$$ P(Y \geq 1) = 1 - P(Y = 0) = 1 - \binom{5}{0} (0.05)^0 (0.95)^5 $$

Calculating the above expression:

$$ P(Y \geq 1) = 1 - 1 \times 1 \times (0.95)^5 $$

$$ P(Y \geq 1) = 1 - 0.7737809375 $$

$$ P(Y \geq 1) = 0.2262190625 $$

So, the probability of observing at least one defective is approximately 0.2262 or 22.62%.

**Question 1:**

The given probability distribution is a Poisson distribution. The expected value (mean) of a Poisson distribution is equal to its parameter, λ. So, the expected value of this distribution is **λ**.

**Proof:**

The expected value E(Y) of a random variable Y is defined as:

$$E(Y) = \sum_{y=0}^{\infty} y \cdot p(y)$$

For a Poisson distribution, this becomes:

$$E(Y) = \sum_{y=0}^{\infty} y \cdot \frac{\lambda^y}{y!} e^{-\lambda}$$

This can be simplified to:

$$E(Y) = \lambda e^{-\lambda} \sum_{y=1}^{\infty} \frac{\lambda^{y-1}}{(y-1)!}$$

The sum in this expression is the sum of probabilities for a Poisson distribution with parameter λ, which sums to 1. Therefore, the expected value is:

$$E(Y) = \lambda$$

**Question 2:**

The number of seedlings in a square yard follows a Poisson distribution with λ = 5. We want to find the probability that none of the 10 regions contain seedlings. 

The probability that a single region contains no seedlings is given by the formula for the Poisson distribution with y = 0:

$$p(0) = \frac{\lambda^0}{0!} e^{-\lambda} = e^{-\lambda}$$

Substituting λ = 5, we get:

$$p(0) = e^{-5}$$

Since the regions are independent, the probability that all 10 regions contain no seedlings is the product of their individual probabilities:

$$p(\text{all empty}) = (e^{-5})^{10} = e^{-50}$$

So, the probability that none of the 10 regions will contain seedlings is **e^-50**.
