# Triangle Inequality Theorem

The Triangle Inequality Theorem states that for any two real numbers `x` and `y`, the following inequality holds true:

$$|x + y| \leq |x| + |y|$$

## Proof

We can prove this theorem by considering two scenarios: when `x + y` is greater than or equal to 0, and when `x + y` is less than 0.

### Scenario 1: `x + y` is greater than or equal to 0

In this scenario, `|x + y|` is equal to `x + y`. Hence, the inequality becomes:

$$x + y \leq |x| + |y|$$

This is obviously true because both `|x|` and `|y|` are non-negative.

### Scenario 2: `x + y` is less than 0

In this scenario, `|x + y|` is equal to `-x - y`. Hence, the inequality becomes:

$$-x - y \leq |x| + |y|$$

If we add `x` to both sides, we get:

$$-y \leq |x| + |y| - x$$

Given that `x + y` is less than 0, it follows that `-y` is greater than or equal to 0, and `|x| + |y| - x` is greater than or equal to `|y| - x`. Therefore, we have:

$$-y \leq |y| - x$$

This is obviously true because `|y|` is non-negative and `-y` is non-positive.

## Conclusion

Since the Triangle Inequality Theorem holds true for both scenarios, we can conclude that it holds true for all real numbers `x` and `y`.
